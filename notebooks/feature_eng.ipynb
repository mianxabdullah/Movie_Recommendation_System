```python
import pandas as pd
movies = pd.read_csv('..\\data\\processed\\movies_clean.csv')
ratings = pd.read_csv('..\\data\\processed\\ratings_clean.csv')

#calculate average rating for each movie
avg_rating = (
    ratings.groupby('movieId')['rating']
    .mean()
    .reset_index()
)

avg_rating.rename(
    columns={'rating': 'avg_rating'},
    inplace=True
)

print("avg_rating.shape: ",avg_rating.shape)
print("Head of Data:\n",avg_rating.head())

```

```python
#calculate rating count for each movie
rating_count = (
    ratings
    .groupby('movieId')['rating']
    .count()
    .reset_index()
)

rating_count.rename(
    columns={'rating': 'rating_count'},
    inplace=True
)

print("rating_count.shape: ",rating_count.shape)
print("Head of Data:\n",rating_count.head())

```

```python
#merge avg_rating and rating_count on movieId to create movie_stats dataframe
movie_stats = avg_rating.merge(
    rating_count,
    on='movieId'
)

print("movie_stats.shape: ",movie_stats.shape)
print("Head of Data:\n",movie_stats.head())
```

```python
#extract genre information as binary columns 
genre_df = movies['genres'].str.get_dummies(sep='|')
genre_df['movieId'] = movies['movieId']
print("genre_df.shape: ",genre_df.shape)
print("Head of Data:\n",genre_df.head())
```

```python
#extract year from title and convert to numeric , Toy Story (1995)
movies['year'] = (
    movies['title']
    .str.extract(r'\((\d{4})\)')
)

movies['year'] = pd.to_numeric(
    movies['year'],
    errors='coerce'
)
movies['year'] = movies['year'].astype('Int64')

#handling missing values in year column by replacing with median of year column
print("movies['year'].isnull().sum(): ",movies['year'].isnull().sum())
movies['year']= movies['year'].fillna(int(movies['year'].median()))
print("movies['year'].isnull().sum(): ",movies['year'].isnull().sum())

print("movies.shape: ",movies.shape)
print("Head of Data:\n",movies.head())
```

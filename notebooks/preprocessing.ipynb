```python
import pandas as pd
movies = pd.read_csv('../data/raw/movies.csv')

print("movies.shape: ",movies.shape)
print("Head of Data:\n",movies.head())
print("Data Info:\n",movies.info())
print("Missing Values:\n",movies.isnull().sum())
print("Duplicate Values:\n",movies.duplicated().sum())
#if duplicate exists movies.drop_duplicates(inplace=True)
print("No of Unique movies",movies["movieId"].nunique())

movies.to_csv(
    "../data/processed/movies_clean.csv",
    index=False
)
```

```python
ratings = pd.read_csv("../data/raw/ratings.csv")
print("data shape:", ratings.shape)
print("first few rows:")
print(ratings.head())
print("missing values:")
print(ratings.isnull().sum())
print("duplicate rows:")
print(ratings.duplicated().sum())
#if duplicate exists ratings.drop_duplicates(inplace=True)
ratings.drop_duplicates(inplace=True)
print("all movie IDs are valid:")
print(ratings['movieId'].isin(movies['movieId']).all())
ratings = ratings.drop(columns=['timestamp'])
print("rating statistics:")
print(ratings['rating'].describe())
print("unique users:")
print(ratings['userId'].nunique())
print("movie rating counts:")
print(ratings.groupby('movieId').size().describe())
ratings.to_csv(
    "../data/processed/ratings_clean.csv",
    index=False
)
```
